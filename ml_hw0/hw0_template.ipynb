```python
# How many malignant cases are there in total?
cancer.target[cancer.target ==  cancer.target_names.tolist().index("malignant")].size
```

```python
# Make a scatterplot of â€˜mean compactnessâ€™ and â€˜mean perimeterâ€™. Try to color the points by
# their label (i.e. malignant or benign). You may find scatter (from matplotlib) or jointplot
# (from seaborn) useful for this.

# matplotlib example
_fig, ax = plt.subplots()

color_map = {'malignant': 'tab:red', 'benign': 'tab:blue'}
color_f = lambda t: color_map[cancer.target_names[t]]
colors = [color_f(x) for x in cancer.target]

ax.scatter(
    df["mean compactness"],
    df["mean perimeter"],
    c=colors,
    alpha=0.5,
    edgecolors='none',
)#, s=scale, label=color,

ax.grid(True)

plt.show()
```

```python
# Make pairwise scatterplots of â€˜mean textureâ€™, â€˜mean perimeterâ€™, â€˜mean compactnessâ€™, and â€˜mean
# radiusâ€™. You may find scatter matrix (from matplotlib) or pairplot (from seaborn) useful
# for this. Do you notice that two of these features seem to be particularly correlated? Why
# might this be?

# A: mean perimeter and mean radius are strongly correlated;
# This is because the perimeter of a circle has a linear relationship to 
# its radius, P = 2 * pi * radius

# seaborn example
# http://seaborn.pydata.org/generated/seaborn.pairplot.html
sns.set(style="ticks", color_codes=True)

labels = pd.Series([cancer.target_names[t] for t in cancer.target])
df["label"] = labels

g = sns.pairplot(
    df,
    vars=["mean texture", "mean perimeter", "mean compactness", "mean radius"],
    hue="label",
    corner=True,
    palette={
        'malignant': 'tab:red',
        'benign': 'tab:blue',
    }
)
```
