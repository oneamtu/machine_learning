```python
# Import libraries:
import numpy as np
import pandas as pd
import seaborn as sns  # optional, but recommended
import matplotlib.pyplot as plt 
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer()
# Check keys:
cancer.keys()
```

```python
# View description:
print(cancer.DESCR)
```

```python
# Check feature names:
cancer.feature_names
```

```python
# And target names:
cancer.target_names
# Note: the target name corresponding to a target (or label) value of i is given by target_names[i].
# Thus we see that 'malignant' is represented by 0 and 'benign' by 1.
```

```python
# Convert data to pandas DataFrame:
df = pd.DataFrame(cancer.data, columns = cancer.feature_names)
print(df.shape)
# Add 'target' and 'diagnosis' keys for convenience 
df['target'] = cancer.target
df['diagnosis'] = np.array([cancer.target_names[i] for i in cancer.target])
df
```

```python
# Viewing the 10th row:
df[9:10]
```

```python
# How many malignant cases do we have?
target = df['target']
q = np.where(target == 0)[0]
len(q)
```

```python
# Plot a scatter plot of 'mean compactness' and 'mean perimeter'
sns.scatterplot(x='mean compactness', y='mean perimeter', hue='diagnosis', data=df)
```

```python
# Alternatively in pure matplotlib:
scatter = plt.scatter(df['mean compactness'], df['mean perimeter'], c=df['target'])
plt.xlabel('mean compactness')
plt.ylabel('mean perimeter')
plt.legend(handles=scatter.legend_elements()[0], labels=list(cancer.target_names))
```

```python
# A pairwise scatter plot of 'mean texture', 'mean perimeter', 'mean compactness', and 'mean radius'
sns.pairplot(df, vars=['mean texture', 'mean perimeter', 'mean compactness', 'mean radius'], hue='diagnosis')
```

We notice that mean perimeter and mean radius are almost perfectly linearly related. This probably just reflects the simple geometric fact that for roughly-circular shapes, perimeter is $2\pi\  \times$ radius!
